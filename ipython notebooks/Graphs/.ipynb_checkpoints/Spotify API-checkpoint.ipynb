```python
sp = spotipy.Spotify(auth=spotify_token)
```

```python
def get_artist(artist_list):
    artist_results_list = []
    for i in artist_list:
        name = [i]
        if name == ["nan"]:
            artist_results_list.append(name)
        else:

            try:
                artist_results = sp.search(q='artist:' + name[0], type='artist')['artists']['items'][0]
                artist_results_list.append([artist_results["followers"]["total"], artist_results["genres"], artist_results["name"], artist_results["popularity"]])
                #print(artist_results)
                #check = artist_results["artists"]
                #artist_results_list += artist_results["followers"]["total"]

            except:
                artist_results_list.append(name)
            
    df_artists = pd.DataFrame(artist_results_list, columns = ["artist_followers","artist_genres", "artist_name", "artist_popularity"])    
    return df_artists
```

```python
list = ["nan", "beatles", "rolling stones", "Ty Segall"]
get_artist(list)
```
